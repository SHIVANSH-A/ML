```python


from ipywidgets import widgets, Layout, Box
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch


model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
device = torch.device('cpu')


def summarize_text(input_text):
    preprocessed_text = input_text.strip().replace('\n', '')
    t5_input_text = 'summarize: ' + preprocessed_text
    tokenized_text = tokenizer.encode(t5_input_text, return_tensors='pt', max_length=512).to(device)
    summary_ids = model.generate(tokenized_text, min_length=30, max_length=120)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

input_text_area = widgets.Textarea(
    placeholder='Enter text here...',
    layout=Layout(width='50%', height='200px')
)


summarize_button = widgets.Button(
    description='Summarize',
    button_style='info',
    button_color = "Orange",
    layout=Layout(width='20%', height='auto')
)


output_text_area = widgets.Textarea(
    placeholder='Summary will appear here...',
    disabled=True,
    layout=Layout(width='50%', height='100px')
)


def on_summarize_button_clicked(b):
    output_text_area.value = summarize_text(input_text_area.value)

summarize_button.on_click(on_summarize_button_clicked)


widgets.VBox([input_text_area, summarize_button, output_text_area], layout=Layout(align_items='center'))
```
